```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

# A) Defining data

```python
x1 = np.random.normal(loc = 0.0, scale = 0.3, size = (100,1))
y1 = np.random.normal(loc = 0.0, scale = 0.3, size = (100,1))
x2 = np.random.normal(loc = 2.0, scale = 0.3, size = (100,1))
y2 = np.random.normal(loc = 2.0, scale = 0.3, size = (100,1))
```

```python
plt.plot(x1, y1, 'o')
plt.plot(x2, y2, 'o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('plot x and y')
plt.legend(['Class1','Class2'])
plt.show()
```

#B) Adaline

```python
# Sorting the data randomly
t1 = np.ones((100,1))
t2 = -np.ones((100,1))
s1 = np.concatenate((x1, y1, t1), axis=1)
s2 = np.concatenate((x2, y2, t2), axis=1)
s = np.concatenate((s1, s2), axis=0)
s = np.random.permutation(s) 
```

```python
#Initialization
W = np.random.random((2,1))  
b = np.random.random() 
alpha = 0.01
```

```python
#Training Adaline
E = []
for epoch in range(20):
  for i in range(200):
    net = np.dot([s[i,0:2]],W) + b
    Err = s[i,2] - net
    W[0,0] = W[0,0] + alpha * Err * s[i,0]
    W[1,0] = W[1,0] + alpha * Err * s[i,1]
    b = b + alpha * Err
    Error = 0.5 * (Err)**2
    E.append(Error)
print("The learned weights are : \n" , W)
```

```python
plt.plot(x1, y1, 'o')
plt.plot(x2, y2, 'o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('plot x and y')
plt.legend(['Class1','Class2'])
line = np.linspace(-100,100)
line = np.reshape(line,(50,1))
out = -(W[0,0] / W[1,0]) * line - (b / W[1,0])
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.plot(line, out)
plt.show()
```

```python
E = np.reshape(E,(4000,))
plt.plot(E)
plt.xlabel('iteration')
plt.ylabel('Error')
plt.title('plot Error curve')
plt.legend(['Error'])
plt.show()
```
