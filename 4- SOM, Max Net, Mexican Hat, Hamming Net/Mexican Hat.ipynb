```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

>Inputs

```python
X = [ 0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08 ]
```

```python
def modify(R1, R2, F, G):
	B1 = - R1 + G
	B2 = - R2 + G
	P1 = R1 + G + 1
	P2 = R2 + G + 1
	if B1 < 0:
		B1 = 0
	if B2 < 0:
		B2 = 0
	if P2 > F:
		P2 = F
	if P1 > F:
		P1 = F
	return B1, B2, P1, P2
```

>Activation Function

```python
def Activation(X):
  if X < 0:
    output = 0
  elif X >= 0 and X < 2:
    output = X
  else:
    output = 2
  return output
```

>Mexican Hat 

```python
def Mexican_hat(X,T_max,R1,R2,C1,C2):
  X1 = X.copy()
  index = []
  for j in range(T_max):
    temp = []
    X_old = X1.copy()
    plt.plot(X_old)
    for i in range(len(X)):
      B1, B2, P1, P2 = modify(R1, R2, len(X), i)
      s1 = C1 * sum((X_old[B1:P1]))
      s2 = C2 * sum(X_old[B2:B1])
      s3 = C2 * sum((X_old[P1:P2]))
      S = s1 - (s2 + s3)
      X1[i] = Activation(S)
      temp.append(X1[i])
    index.append(temp)
  print(pd.DataFrame(index))
  plt.plot(X1)
  plt.legend(['t = 0', 't = 1', 't = 2', 't = 3'])
  plt.grid()
  plt.show()
```

>Part 1- 
>>R1 = 0 & R2 = Inf

```python
T_max = 4
R1 = 0
R2 = 1000
C1 = 1.5
C2 = 0.08
```

```python
Mexican_hat(X,T_max,R1,R2,C1,C2)
```

>Part 2- 
>>R1 = 1 & R2 = 3

```python
T_max = 4
R1 = 1
R2 = 3
C1 = 0.7
C2 = 0.5
```

```python
Mexican_hat(X,T_max,R1,R2,C1,C2)
```
