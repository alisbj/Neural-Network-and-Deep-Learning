```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

#Loading and Visualizing Data

```python
data = pd.read_csv('Question3.csv',header = None)
X1 = data[0]
Y1 = data[1]
plt.plot(X1[0:100], Y1[0:100], 'o')
plt.plot(X1[100:200], Y1[100:200], 'o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('plot x and y')
plt.legend(['Class1','Class2'])
plt.show()
```

#Sorting randomly the data & creat -1 labels

```python
d = np.random.permutation(data) 
X = d[:,0:2]
Y = d[:,1]
T = d[:,2]
T[T==0] = -1
```

#Activation Function


```python
def Activation(X):
  if X >= 0:
    f = 1
  else:
    f = -1
  return f
```

#Algorithm for 3 neruns

```python
V = 0.5 * np.ones((3,1))
b2 = 0.5
W = np.random.rand(2,3)
b1 = np.random.rand(3,1)
alpha = 0.15
Z = np.zeros((3,1))
C = 0
epoch = 0
for ep in range(200):
  if C >= 175:
    epoch = ep + 1
    break
  C = 0
  for i in range(200):
    Z_in = np.dot(W.T,np.reshape(X[i,:],(2,1))) + b1
    P1 = [abs(Z_in[0]),abs(Z_in[1]),abs(Z_in[2])]
    Z[0] = Activation(Z_in[0])
    Z[1] = Activation(Z_in[1])
    Z[2] = Activation(Z_in[2])
    y_in = np.dot(V.T,Z) + b2
    y = Activation(y_in)
    if y == T[i]:
      C = C + 1
      continue
    if (y != T[i]) & (T[i] == 1):
      id = P1.index(min(P1))
      W[:,id] = W[:,id] + alpha * (T[i] - Z_in[id]) * np.reshape(X[i,:],(2,))
      b1[id,0] = b1[id,0] + alpha * (T[i] - Z_in[id])
    if (y != T[i]) & (T[i] == -1):
      for j in range(3):
        if Z_in[j] > 0:
          W[:,j] = W[:,j] + alpha * (T[i] - Z_in[j]) * np.reshape(X[i,:],(2,))
          b1[j,0] = b1[j,0] + alpha * (T[i] - Z_in[j])
print('The number of epochs is : \n',epoch)
print('The accuracy is : \n',(C/200)*100)
```

```python
plt.plot(X1[0:100], Y1[0:100], 'o')
plt.plot(X1[100:200], Y1[100:200], 'o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('plot x and y')
plt.legend(['Class1','Class2'])
l1 = np.linspace(-100,100)
out1 = -(W[0,0]/W[1,0])*l1 - (b1[0,0]/W[1,0])
out2 = -(W[0,1]/W[1,1])*l1 - (b1[1,0]/W[1,1])
out3 = -(W[0,2]/W[1,2])*l1 - (b1[2,0]/W[1,2])
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot(l1, out1)
plt.plot(l1, out2)
plt.plot(l1, out3)
plt.show()
```
